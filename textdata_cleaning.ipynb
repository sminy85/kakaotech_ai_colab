````
# 불용어 제거 -- 한국어
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from konlpy.tag import Okt

okt = Okt()

exampleTit = lower_tit
exampleSub = lower_sub
stop_words = "은 는 이 가 을 으로 이다 인다 를 한 와 에서 으로 로 에 는 다 했다 된다 했다"

stop_words = set(stop_words.split(' '))
tit_tokens = okt.morphs(exampleTit)
sub_tokens = okt.morphs(exampleSub)

tit_token = [word for word in tit_tokens if not word in stop_words]
sub_token = [word for word in sub_tokens if not word in stop_words]

print('불용어 제거 전 :',tit_tokens)
print('불용어 제거 후 :',tit_token)
print('불용어 제거 전 :',sub_tokens)
print('불용어 제거 후 :',sub_token)
````

````
# cleaned_data.txt 파일로 저장하기
cleaned_tit = ' '.join(tit_token)
cleaned_sub = ' '.join(sub_token)
print(cleaned_tit)
print(cleaned_sub)
````

````
# 클리닝 된 텍스트 데이터 cleaned_data.txt 파일로 저장
with open('cleaned_data.txt','w',encoding='utf-8') as f:
    f.write(cleaned_tit)
    f.write(cleaned_sub)
````

````
import re

# 특수문자 제거
new_title = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\]^_\`{|}~\\\\]', '', title)
print(new_title)
new_cont = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\]^_\`{|}~\\\\]', '', cont)
print(new_cont)
````

````
# 불용어 제거
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
tokens = word_tokenize(lower_title)
filtered_title = [token for token in tokens if token not in stop_words]
#filtered_title = ' '.join(filtered_tokens)
print(filtered_title)

tokens = word_tokenize(lower_cont)
filtered_cont = [token for token in tokens if token not in stop_words]
#filtered_cont = ' '.join(filtered_tokens)
print(filtered_cont)


````

````
from spellchecker import SpellChecker

spell = SpellChecker()
spellcheck_title = []
spellcheck_cont = []

for word in filtered_title:
    spellcheck_title.append(spell.correction(word))

for word in filtered_cont:
    correction = spell.correction(word)
    if correction is not None:  # Check if a correction was found
        spellcheck_cont.append(correction)

print(spellcheck_title)
print(spellcheck_cont)
````

````
# 클리닝 된 텍스트 데이터
cleaned_title = ' '.join(spellcheck_title)
cleaned_cont = ' '.join(spellcheck_cont)
print(cleaned_title)
print(cleaned_cont)
````

````
# 어간 추출하기
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()

tokenized_cleanedtitle = word_tokenize(cleaned_title)
tokenized_cleanedcont = word_tokenize(cleaned_cont)

print('어간 추출 전 :', tokenized_cleanedtitle)
print('어간 추출 후 :',[stemmer.stem(word) for word in tokenized_cleanedtitle])
print('어간 추출 전 :', tokenized_cleanedcont)
print('어간 추출 후 :',[stemmer.stem(word) for word in tokenized_cleanedcont])


````

````
# 원형 복원(표제어 복원)
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatizer_title = word_tokenize(cleaned_title)
lemmatizer_cont = word_tokenize(cleaned_cont)

lemmatizered_title = [lemmatizer.lemmatize(word, 'v') for word in lemmatizer_title]
lemmatizered_cont = [lemmatizer.lemmatize(word, 'v') for word in lemmatizer_cont]

print('표제어 추출 전 :',lemmatizer_title)
print('표제어 추출 후 :',lemmatizered_title)
print('표제어 추출 전 :',lemmatizer_cont)
print('표제어 추출 후 :',lemmatizered_cont)
````

````
# 정규화 된 텍스트 데이터
normalized_title = ' '.join(lemmatizered_title)
normalized_cont = ' '.join(lemmatizered_cont)
print(normalized_title)
print(normalized_cont)
````

````
# 정규화 된 텍스트 데이터 cleaned_data.txt 파일로 저장
with open('normalized_data.txt','w',encoding='utf-8') as f:
    f.write(normalized_title)
    f.write(normalized_cont)
````

2.   upper() : 문자열을 대문자로 변환하기

### 어간 추출(Stemming) 또는 원형 복원(Lemmatization)

1.   어간 추출(Stemming)

    형태학적 분석을 단순화한 버전

    어간 추출 후 나오는 결과 단어가 사전에 존재하지 않을 수 있음

    Porter 알고리즘

      ```
      from nltk.stem import PorterStemmer
      from nltk.tokenize import word_tokenize

      stemmer = PorterStemmer()

      sentence = 'Example sentence for stemming. Hi kakaotech bootcamp!'
      tokenized_sentence = word_tokenize(sentence)

      print('어간 추출 전 :', tokenized_sentence)
      print('어간 추출 후 :',[stemmer.stem(word) for word in tokenized_sentence])

      ```

    NLTK에서는 포터 알고리즘 외에도 랭커스터 스태머(Lancaster Stemmer) 알고리즘을 지원
      ```
      from nltk.stem import LancasterStemmer

      lancaster_stemmer = LancasterStemmer()

      print('랭커스터 스테머의 어간 추출 후:',[lancaster_stemmer.stem(w) for w in words])
      ```


2.    원형 복원/표제어 추출(Lemmatization)

      표제어란 표제어, 또는 기본 사전형 단어 정도의 의미를 갖는다.

      표제어 추출 -> 형태학적 파싱 먼저 진행하기(형태소(어간/접사)로부터 단어 만들기)

        ```
        from nltk.stem import WordNetLemmatizer
        from nltk.tokenize import word_tokenize

        lemmatizer = WordNetLemmatizer()

        sentence = 'Example sentence for stemming. Hi kakaotech bootcamp!'
        words = word_tokenize(sentence)

        print('표제어 추출 전 :',words)
        print('표제어 추출 후 :',[lemmatizer.lemmatize(word) for word in words])
        ```
