# About Dataset

The dataset consists of 100+ patient records. Each record contains the following information:

Age: The patient's age, in years.

Gender: The patient's gender, male or female.

BMI: The patient's body mass index (BMI), a measure of weight relative to height.

Blood pressure: The patient's blood pressure, in mmHg.

FBS: The patient's fasting blood sugar, in mg/dL.

HbA1c: The patient's hemoglobin A1c, a measure of blood sugar control over the past 3 months.

Family history of diabetes: Whether the patient has a family history of diabetes.

Smoking: Whether the patient smokes.

Diet: Whether the patient has a poor or healthy diet.

Exercise: Whether the patient exercises regularly.

Diagnosis: The patient's diagnosis, either diabetes or no diabetes.

```python
# CSV load
data = pd.read_csv('Diabetes_Classification.csv')

# 데이터의 처음 몇 줄을 출력하여 구조 확인
print("#"*10)
print("데이터의 처음 몇 줄을 출력하여 구조 확인")
print(data.head())

# 데이터의 각 컬럼에 대한 정보 확인
print("#"*10)
print("데이터의 각 컬럼에 대한 정보 확인")
print(data.info())
```

# 막대 그래프를 생성하여 범주형 데이터의 분포를 시각화합니다.

```python
data['Blood Pressure'].head()
```

```python
normal = 0
high = 0
low = 0
for col in data["Blood Pressure"]:
  if col == 'Normal':
    normal += 1
  elif col == 'High':
    high += 1
  elif col == 'Low':
    low += 1

print(normal)
print(high)
print(low)
```

```python
# 1. 막대그래프
bloodPressure = ['Normal', 'High', 'Low']
cases = [normal, high, low]

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.bar(bloodPressure, cases, color='skyblue', label='Blood Pressure')  # 막대그래프 생성
plt.title('Blood Pressure')  # 제목 추가
plt.xlabel('cases')  # x축 레이블 추가
plt.ylabel('How many')  # y축 레이블 추가
plt.legend()  # 범례 추가
plt.grid(True)  # 그리드 추가

# 그래프를 JPEG 형식으로 저장
plt.savefig('bargraph_bloodpressure.jpg', format='jpeg')
plt.show()  # 그래프 출력
```

# 선 그래프를 생성하여 시간에 따른 데이터의 변화를 시각화합니다.

```python
underweight = 0
normalweight = 0
pre_obesity = 0
obesity1 = 0
obesity2 = 0
obesity3 = 0

for i in range(len(data['BMI'])):
  if data['BMI'][i] <= 18.5:
    underweight += 1
  elif data['BMI'][i] <= 22.9:
    normalweight += 1
  elif data['BMI'][i] <= 24.9:
    pre_obesity += 1
  elif data['BMI'][i] <= 29.9:
    obesity1 += 1
  elif data['BMI'][i] <= 34.9:
    obesity2 += 1
  else:
    obesity3 += 1
```

```python
# 데이터 정의
data = {'BMI': ['underweight', 'normalweight', 'pre-obesity', 'obesity1', 'obesity2', 'obesity3'], 'cases': [underweight, normalweight, pre_obesity, obesity1, obesity2, obesity3]}
df = pd.DataFrame(data)
```

```python
# 2. 선그래프
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
sns.lineplot(x='BMI', y='cases', data=df, marker='o', label='Cases')
plt.title('Body mass index')  # 제목 추가
plt.xlabel('BMI')  # x축 레이블 추가
plt.ylabel('Number of cases')  # y축 레이블 추가
plt.legend()  # 범례 추가
plt.grid(True)  # 그리드 추가

# 그래프를 JPEG 형식으로 저장
plt.savefig('linegraph_BMI.jpg', format='jpeg')
plt.show()
```

# 산점도를 생성하여 두 변수 간의 상관관계를 시각화합니다.

```python
# CSV load
data = pd.read_csv('Diabetes_Classification.csv')

BMI = []
for i in range(len(data['BMI'])):
  BMI.append(float(data['BMI'][i]))

HbA1c = []
for i in range(len(data['HbA1c'])):
  HbA1c.append(float(data['HbA1c'][i]))
```

```python
data = {
    'x': BMI,
    'y': HbA1c,
}
df = pd.DataFrame(data)
```

```python
# 3. 산점도
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
sns.scatterplot(x='x', y='y', data=df)
plt.title('Scatter Plot of BMI & HbA1c')  # 제목 추가
plt.xlabel('BMI')  # x축 레이블 추가
plt.ylabel('HbA1c')  # y축 레이블 추가
plt.grid(True)  # 그리드 추가

# 그래프를 JPEG 형식으로 저장
plt.savefig('scatter_plot_BMIxHbA1c.jpg', format='jpeg')
plt.show()
```

# 히스토그램을 생성하여 데이터의 분포를 시각화합니다.

```python
# 4. 히스토그램
# CSV load
data = pd.read_csv('Diabetes_Classification.csv')

HbA1c = []
for i in range(len(data['HbA1c'])):
  HbA1c.append(float(data['HbA1c'][i]))

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.hist(HbA1c, bins=10, edgecolor='black')
plt.title('HbA1c')
plt.xlabel('HbA1c')
plt.ylabel('Frequency')
plt.hist(HbA1c, bins=30, edgecolor='black', label='bins=30')
plt.legend()

# 그래프를 JPEG 형식으로 저장
plt.savefig('hist_HbA1c.jpg', format='jpeg')
plt.show()
```

# 파이차트로 범주별 구성 비율을 원형으로 표현합니다.

```python
yes = 0
no = 0

for col in data['Family History of Diabetes']:
  if col == 'Yes':
    yes += 1
  elif col == 'No':
    no += 1

print(yes)
print(no)
```

```python
labels = ['Yes', 'No']
sizes = [yes, no]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # 첫 번째 조각을 약간 떨어뜨리기
```

```python
# 5. 파이차트
plt.figure(figsize=(8, 8))  # 그래프 크기 설정
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Family History of Diabetes')
plt.axis('equal')  # 파이차트를 원형으로 유지

# 그래프를 JPEG 형식으로 저장
plt.savefig('pie_FamilyHistoryofDiabetes.jpg', format='jpeg')
plt.show()
```

# 박스플롯을 생성하여 데이터의 분포와 이상치를 시각화합니다.

```python
# 6. 박스플롯
plt.figure(figsize = (10, 10))
plt.boxplot([data['FBS'], data['HbA1c'], data['BMI']], whis=0.5)
plt.title('Box Plot of FBS, HbA1c, BMI')
plt.xticks([1, 2, 3], ['FBS', 'HbA1c', 'BMI'])
plt.ylabel('Values')
# 그래프를 JPEG 형식으로 저장
plt.savefig('boxplot_FBSxHbA1cxBMI.jpg', format='jpeg')
plt.show()
```
